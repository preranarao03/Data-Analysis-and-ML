```python
import pandas as pd
# UNDERSTANDING YOUR DATA

# 7 BASIC QUESTIONS TO ASK WHEN WE RECIEVE THE DATA:
```

```python
df = pd.read_csv("train.csv")
```

```python
# HOW BIG IS THE DATA?

# df.shape will tell us the number of rows and columns in the dataset

df.shape # 891 rows, 12 cols
```

```python
# HOW DOES THE DATA LOOK LIKE ?

# df.head() tells us the first few rows 

df.head() # by default the first 5 rows will be printed
```

```python
# to print any number of random rows use .sample (specify the numbe of rows you want to see)

df.sample(5) # will produce any 5 random rows
```

```python
# WHAT IS THE DATA TYPE OF THE COLS? 

# you will info about the columnn using .info()

df.info()

# tells us the number of non-null entries, and dtype of column
```

```python
# ARE THERE ANY MISSING VALUES ?

# you should know this beforehand 
# to get to know the number of missing values in each column do:

df.isnull().sum() # this will give total number of null values in each columns
```

```python
# HOW DOES DATA LOOK MATHEMATICALLY ?

# give a mathematical summary of the numerical columns

df.describe()
```

```python
# ARE THERE ANY DUPLICATE VALUES ?

df.duplicated().sum() # tells us the total number of duplicated values in each column
```

```python
# HOW IS THE CORRELATION BETWEEN THE COLS ?

# if one value is increasing or decreasing what is its effect on another value

# the input columns that do not have a large impact on the output columns may be removed to improve efficiency

df.corr() # tells us the correlation between each column and the other columns

# if positive correlation -> directly proportional
# if negative correlation -> inversly proportional
```
